```python
def get_mean(priceList, time):
    total_price = 0
    total_days = 0
    for total_days in range(len(priceList) - 1):
        total_price += priceList[total_days][0]
        total_days += 1
    mean = total_price/total_days
    return mean

def get_interval(priceList, time):
    total_price = 0
    total_days = 0
    max_price = 0
    min_price = 999999999
    for total_days in range(len(priceList) - 1):
        if priceList[total_days][0] > max_price:
            max_price = priceList[total_days][0]
        if priceList[total_days][0] < min_price:
            min_price = priceList[total_days][0]
        total_price += priceList[total_days][0]
        total_days += 1
    mean = total_price/total_days
    price_interval = []
    price_interval.append(min_price*.25 + mean*.75)
    price_interval.append(max_price*.25 + mean*.75)
    return price_interval

# function to calculate bollinger bands
# prices - pandas df
# time_period - int, specifies lookback period
def buy_token(balance, price, prev_bought_price):
    if prev_bought_price == 0:
        return
    balance -= price
    prev_bought_price = price

def sell_token(balance, price, prev_sold_price):
    if prev_sold_price == 0:
        return
    balance += price
    prev_sold_price = price

```

```python
def main(priceList, company):
    curr_time = 0
    prev_bought_price_mean = 0
    prev_sold_price_mean = 0
    prev_bought_price_interval = 0
    prev_sold_price_interval = 0
    mean_balance = 0
    interval_balance = 0
    for curr_time in range(len(priceList) - 1):
        curr_price = priceList[curr_time][0]
        mean_price = get_mean(priceList, curr_time)
        mean_interval = get_interval(priceList, curr_time)

        if prev_bought_price_interval != 0:
            interval_balance += curr_price
        if prev_sold_price_interval != 0:
            interval_balance -= curr_price
        prev_bought_price_interval = 0
        prev_sold_price_interval = 0

        if prev_bought_price_mean != 0:
            mean_balance += curr_price
        if prev_sold_price_mean != 0:
            mean_balance -= curr_price
        prev_bought_price_mean = 0
        prev_sold_price_mean = 0

        if curr_price > mean_price:
            mean_balance += curr_price
            prev_sold_price_mean = curr_price
        if curr_price < mean_price:
            mean_balance -= curr_price
            prev_bought_price_mean = curr_price
        if curr_price > mean_interval[1]:
            interval_balance += curr_price
            prev_sold_price_interval = curr_price
        if curr_price < mean_interval[0]:
            interval_balance -= curr_price
            prev_bought_price_interval = curr_price

    mean_balance = mean_balance + prev_bought_price_mean - prev_sold_price_mean
    interval_balance = interval_balance + prev_bought_price_interval - prev_sold_price_interval
    print(company + " mean balance is: " + str(mean_balance))
    print(company + " interval balance is: " + str(interval_balance))      
```
